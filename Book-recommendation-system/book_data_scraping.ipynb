```python
import requests
from bs4 import BeautifulSoup

def scrape_goodreads_books(url):
    # Send a GET request to the provided URL
    response = requests.get(url)
    
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all book containers on the page
    book_containers = soup.find_all('tr', itemtype='http://schema.org/Book')
    
    # Iterate over each book container
    for container in book_containers:
        # print the whole container
        # print(container)
        
        # Extract book title
        title = container.find('a', class_='bookTitle').text.strip()
        
        # Extract book author(s)
        authors = [author.text for author in container.find_all('span', itemprop='name')]
        
        # Extract book rating
        rating = container.find('span', class_='minirating').text.strip()
        
        # Extract the image-url field
        image_url = container.find('img', itemprop='image')['src']
        
        # Print the scraped book data
        print(f'Title: {title}')
        print(f'Authors: {", ".join(authors)}')
        print(f'Rating: {rating}')
        print(f'Image URL: {image_url}')

        print('---')

# URL of the Goodreads page to scrape
url = 'https://www.goodreads.com/search?q=python'

# Call the function to scrape the book data
scrape_goodreads_books(url)
```
